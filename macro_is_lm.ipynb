```sage
# Good market Variables
Y, R = var('Y, R')
C, I, G = var('C, I, G')
c0, c1, c2, i0, i1, i2 = var('c0, c1, c2, i0, i1, i2')

# Goods market Equilibrium condition
Con = C == c0 + c1*Y + c2*R
Inv = I == i0 + i1*Y + i2*R
IS = Y == C + I + G

# Money/Bond market equations
L, P, Ms = var('L, P, Ms')
l0, l1, l2 = var('l0, l1, l2')

# Money market equations
LP = L == l0 + l1*Y + l2*R
LM = Ms == P * L
```

```sage
Y_eq, R_eq = solve([IS.subs([Con, Inv]),LM.subs(LP)],[Y,R])[0]
```
